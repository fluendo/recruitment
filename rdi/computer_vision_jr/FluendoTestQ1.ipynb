## Q1(1) - First Question

**What do you think occurred during this model development (training & evaluation)?** 

Deep learning methods used for object detection and tracking extract important features and representations by themselves. Convolutional neural networks, which are the primary architecture for Computer Vision tasks, possess several layers with adjustable weights that map pictures through a succession of operations that extract information that is more and more specific the deeper you go in the architecture. Through powerful mechanisms such as back-propagation, these operations can be automatically tuned and optimized for a specific task.

In this specific case of study, a deep learning algorithm was mostly trained to find a ball in the picture of a football field, repeating this operation on each frame of a video while keeping information of at least one frame ago*. The features extraction customized during training most likely learnt to distinguish the ball as a round object with great contrast with the green background thanks to its whitish color. This might as well have been the reason why the AI camera operator later deployed by Scottish soccer team Inverness Caledonian Thistle FC malfunctioned during the match against Ayr United. Since the bold head of one of the linemans resembled the ball based on the learnt features list, the AI could easily be confused during inference therefore identifying the bold head as the ball on several occasions.

This is only one of many possible problems that AI-driven algorithms can face during object tracking. Another common problem is occlusion, an interference phenomenon that can get AI trackers confused whenever multiple objects come close and then get separated again. There can also be generalization errors whenever the target object varies in shape or size during the video or images sequence.

**Online tracking algorithms are usually chosen for real-time applications, taking information from both the current and the last frame of a video. Offline tracking algorithms take a stack of frames in order to make predictions and due to the big load required, their applications are mostly into Video Querying or Video Analysis.*


## Q1(2) - Second Question

**How would you fix this behavior? Please provide at least 2 options explaining their pros and drawbacks**

To begin with, if it has not been done already, it could be a good idea to perform Multiple Object Tracking (MOT) instead of Single Object Tracking (SOT). If a deep learning algorithm was trained to identify not only the ball but also the players at the same time, then the features extraction could have easily helped to distinguish between people and the ball. If that were the case, the lineman could never have been mistaken with the ball, which was what the AI-camera operator was mostly focused on.

Drawbacks of this approach could be that new problems could appear. Although the feature extraction could be improved, players and linemen could get confused. Moreover, individuals could get confused with each other during identification in close distance situations where occlusion could arise.

Another logical step to take would be to work on the predictive tracking system itself. To not rely completely on feature extraction, a better estimation of possible trajectories based on passed positions could help. Either this should be included in the algorithm or it should be given a bigger weight. Maybe it could be possible to consider more than just the last and current frame, as in online tracking, but also some more frames to more accurately predict speed. Perhaps bounds on speed and location could be adaptively learnt, too, given the system some long-term memory.

Special care should be taken while considering these options, though, in order for the algorithm to still be light and fast, since this is most probably a priority in this real-time application. Including these improvements in the tracking predictions might not be easy and the AI model might need to be changed altogether, which could take way too much time, so research should be carried out in advance to determine whether this could be possible or not.

Additional custom-made candidate solutions could include some data preprocessing, such as a frame-by-frame identification of an area of interest that could be reduced to the active field, a part of which the linesman is not, or even a moving area that surrounds the object. Nevertheless, this could prove difficult and time-consuming, not generalizable and with a heavy impact on the inference time.


## Q1(3) - Third Question

I believe DeepSORT is an example of a deep learning-based algorithm that could be used here. It is a popular object tracking algorithm that introduces deep learning into the SORT algorithm (Simple Online Real-time Tracking) by adding an appearance descriptor to reduce identity switches. 

In the SORT algorithm, objects are first detected on a frame and then propagated to the next frame by a constant velocity model estimation. After this is done, data association between objects and IDs is carried out by computing IOU distance (Intersection-Over-Union) between all pairs of boxes. In DeepSORT, a better association metric that combines both motion and appearance descriptors. To achieve this, a well-discriminating feature embedding is trained offline just before implementing tracking.
